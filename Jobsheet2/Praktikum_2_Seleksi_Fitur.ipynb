0. Load Library

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
import io
```

1. Load Data

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['Titanic-Dataset-fixed.csv'])) #load dataset

dpath = 'Titanic-Dataset-fixed.csv'
df = pd.read_csv(dpath)
df.head()
```

2. Slice Data

```python
df = df[['Survived', 'Pclass', 'Age', 'Sex', 'Cabin']]
df.head()
```

3. Encoding

```python
le = LabelEncoder() # membuat objek dari LabelEncoder
df['Sex'] = le.fit_transform(df['Sex']) # proses encoding
df['Cabin'] = le.fit_transform(df['Cabin']) # proses encoding
```

4. Verifikasi Hasil

```python
df.head()
```

5. Standarisasi

```python
std = StandardScaler()
df['Age'] = std.fit_transform(df[['Age']])
```
